```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
random.seed(101)
```

```python
#lets first assume that we have data belonging to 2 different classes
#one is positive class represented as '+'
#second class is negative class represented as '-'
x_train=[]
y_train=[]
for i in range(1,5):
    x_train.append(random.randint(0,4))
    y_train.append(random.randint(4,10))

for i in range(1,5):   
    x_train.append(random.randint(8,12))
    y_train.append(random.randint(0,3))    
    
    
```

```python
#now lets plot those two classes and
#lest see overall positions of the two different classes
for i in range(len(x_train)):
    if i<4:
        plt.scatter(x_train[i],y_train[i],marker='+',linewidth=2)
    else:
        plt.scatter(x_train[i],y_train[i],marker='_',linewidth=2)
  
```

```python
#we will now be aiming at forming a seperator which could seperate these two classes
#so we choose one point from each class such that two classes are close to each other
#so our required points are  (9,2) and (2,7) and so these points will lie on our support vectors of SVM
#seperator we required will be perpendicular to these two points and will pass through the mid point of these two so our 
#mid point=(7/2,5/2)
#seperator eqation will be (y-(5/2))=( -(9-2)/(2-7) )(x-(7/2))
#i.e  on solving we will get 7x-5y-12=0
#hence our f(x) will become (7*x-12)/5
```

```python
#lets plot our seperator equation
x=np.array(range(100))
plt.plot(x,(7*x-12)/5, color='red')
```

```python
#now if we plot our points along with the classifier 
#we can see that our classifier is properly seperating the training points
#two classes in the following graph
for i in range(len(x_train)):
    if i<4:
        plt.scatter(x_train[i],y_train[i],marker='+',linewidth=4)
    else:
        plt.scatter(x_train[i],y_train[i],marker='_',linewidth=4)
x=np.array(range(15))
plt.plot(x,(7*x-12)/5, color='red')
```

```python
#so lets test our seperator on test points :-
x=np.array(range(15))
plt.plot(x,(7*x-12)/5, color='red')
#so lets test our seperator with other test points
x_test=[]
y_test=[]
for i in range(1,20):      
    x_test.append(random.randint(0,4))
    y_test.append(random.randint(4,10))

for i in range(1,20):       
    x_test.append(random.randint(8,12))
    y_test.append(random.randint(0,3))
for i in range(len(x_test)):
    if i<20:
        plt.scatter(x_test[i],y_test[i],marker='+',linewidth=2)
    else:
        plt.scatter(x_test[i],y_test[i],marker='_',linewidth=2)
```
